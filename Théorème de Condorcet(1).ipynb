Théorème de Condorcet
====
Axe x : $p$ = Fiabilité moyenne des participants (= probabilité moyenne individuelle des participants à croire en une information correcte)

Axe y : $N$ = Nombre de participants

Axe z : $P_N$ = Fiabilité du groupe.

$m$ = $\frac{N}{2}$ = Nombre de partipants requis pour avoir une majorité

Théorème du jury de Condorcet : 
$$P_N = \sum_{i=m}^{N}
\bigg(\frac{N!}{(N-i)!i!}\bigg)(p)^i(1-p)^{N-i}$$

```python
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode()
from plotly.graph_objs import *
import math

Ps = []
x = []
y = []

for N in range(4, 501): # De 4 à 500 participants
    m = int(N/2) # Calculer m
    P_p = []
    y.append(N)
    for p in range(1, 100): # D'une fiabilité nulle à quasi-totale
        P = 0
        for i in range(m, N):
            P += (math.factorial(N)/(math.factorial(N-i)*math.factorial(i)))*((p/100)**i)*((1-(p/100))**(N-i)) # Calculer le théorème de Condorcet
        P_p.append(P)
    Ps.append(P_p)

for i in range(1, 100):
    x.append(i/100)

# Afficher graphique
data = [Surface(z=Ps, x=x, y=y, colorscale='Viridis')]
layout = Layout(height=1000,
                title='Théorème du jury de Condorcet avec m = N/2',
                scene=Scene(
                            xaxis=dict(title='p = fiabilité moyenne des participants', dtick=0.25),
                            yaxis=dict(title='N = nombre de participants', dtick=125),
                            zaxis=dict(title='P_N = fiabilité globale du groupe')))
fig = Figure(data=data, layout=layout)
iplot(fig)
```
