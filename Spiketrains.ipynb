In the brain, neurons discharges spikes of electric potential. Thoses spikes occurs partly randomly. Let's simulate a neuron's discharging of electric energy using an array where 0 signifies that the neuron isn't discharging a spike and where 1 signifies he's discharging a spike. Spikes occurs randomly with a probablity of 0.25.

```python
import random
spikes = []
x = np.linspace(0, 1000, 1000)

for i in range(1000):
    rand = random.random()
    if rand <= 0.25:
        spikes.append(1)
    else:
        spikes.append(0)

trace = Heatmap(z=[spikes])
data = [trace]
layout = Layout(
    title='Spike train',
    xaxis=dict(title='Time (ms)'),
    yaxis=dict(title='Presence/Absence of a spike'))
fig = Figure(data=data, layout=layout)

iplot(fig)
```

Integrate-and-Fire neuron
----
Let's simulate the creation of action potentials by a neuron to see how spike trains emerges. To do so, we need to simulate the voltage variation in a neuron's membrane. Let's give an intensity $I$ = 1nA to our neuron. The membrane capacitance is $C$ = 1nF, the membrane conductance is $g_L$ = 0.1$\mu$S and the reversal potential is $E_L$ = 70mV.

The evolution of the voltage is given by :

$$C\frac{dV(t)}{dt}=g_L(E_L-V(t))+I$$

So to compute the voltage value $V(t+\Delta t)$ we have to apply this formula :

$$V(t+\Delta t)=V(t)+\frac{g_L(E_L-V(t))+I}{C}$$

```python
gL = 0.1
EL = -70
I = 1
C = 1
Vs = [EL]
deltat = 1

for i in range(99):
    Vs.append(Vs[i]+((gL*(EL-Vs[i])+I)/C)*deltat)

x = np.linspace(0, 100, 100)

trace = Scatter(x=x, y=Vs)
data = [trace]

layout = Layout(
    title='Variation of potential in a neuron',
    xaxis=dict(title='Time (ms)'),
    yaxis=dict(title='Electric potential (mV)'))
fig = Figure(data=data, layout=layout)

iplot(fig)
```

We observe that in this model, the potential increases then saturates at a certain value.

Let's try to variate the intensity $I$ to see how it affects the variation of potential.

```python
gL = 0.1
EL = -70
Is = [1, 0.5, 2]
C = 1
Vs = [EL]
variations = []
deltat = 1

for I in Is:
    for i in range(99):
        Vs.append(Vs[i]+((gL*(EL-Vs[i])+I)/C)*deltat)
    variations.append(Vs)
    Vs = [EL]

x = np.linspace(0, 100, 100)

trace1 = Scatter(x=x, y=variations[0], name="I = 1")
trace2 = Scatter(x=x, y=variations[1], name="I = 0.5")
trace3 = Scatter(x=x, y=variations[2], name="I = 2")
data = [trace1, trace2, trace3]

layout = Layout(
    title='Variation of potential in a neuron with different intensities',
    xaxis=dict(title='Time (ms)'),
    yaxis=dict(title='Electric potential (mV)'))
fig = Figure(data=data, layout=layout)

iplot(fig)
```

We observe here that different intensity gives us different thresholds. Lower is the intensity, lower is the threshold of potential. Higher is the instensity, higher is the threshold.

What happens if we change the $\Delta t$ ? 

```python
gL = 0.1
EL = -70
I = 1
C = 1
Vs = [EL]
variations = []
deltats = [1, 0.5, 2]

for deltat in deltats:
    for i in range(99):
        Vs.append(Vs[i]+((gL*(EL-Vs[i])+I)/C)*deltat)
    variations.append(Vs)
    Vs = [EL]

x = np.linspace(0, 100, 100)

trace1 = Scatter(x=x, y=variations[0], name="Deltat = 1")
trace2 = Scatter(x=x, y=variations[1], name="Deltat = 0.5")
trace3 = Scatter(x=x, y=variations[2], name="Deltat = 2")
data = [trace1, trace2, trace3]

layout = Layout(
    title='Variation of potential in a neuron with differents timesteps',
    xaxis=dict(title='Time (ms)'),
    yaxis=dict(title='Electric potential (mV)'))
fig = Figure(data=data, layout=layout)

iplot(fig)
```

We observe that the potential threshold is more quickly reached with an higher $\Delta t$. On the contrary, wih a lower $\Delta t$, the threshold is slower reached.

Now to transform this function into a spikes generating function we need to add a potential threshold $V_th$ = -63mV. When this threshold is reached, the neuron discharges a spike and return to its reversal potential $E_L$.

```python
gL = 0.1
EL = -70
I = 1
C = 1
Vs = [EL]
deltat = 1
Vth = -63
spikescounter = 0

for i in range(99):
    V = Vs[i]+((gL*(EL-Vs[i])+I)/C)*deltat
    if Vs[i] < Vth:
        Vs.append(V)
    else:
        spikescounter += 1
        Vs.append(EL)
x = np.linspace(0, 100, 100)
trace = Scatter(x=x, y=Vs)
data = [trace]

layout = Layout(
    title='Variation of potential in a neuron with Vth',
    xaxis=dict(title='Time (ms)'),
    yaxis=dict(title='Electric potential (mV)'))
fig = Figure(data=data, layout=layout)

iplot(fig)
print("Number of spikes : " + str(spikescounter))
```

If we try different intensities, it changes the number of spikes because the intensity changes the speed of potential increasing and the maximum potential threshold of the neuron so that the threshold $Vth$ is reached more or less quickly.

```python
gL = 0.1
EL = -70
Is = [1, 0.8, 2]
C = 1
Vs = [EL]
deltat = 1
Vth = -63
spikescounter = 0
variations = []
spikecounts = []

for I in Is:
    for i in range(99):
        V = Vs[i]+((gL*(EL-Vs[i])+I)/C)*deltat
        if Vs[i] < Vth:
            Vs.append(V)
        else:
            spikescounter += 1
            Vs.append(EL)
    variations.append(Vs)
    spikecounts.append(spikescounter)
    Vs = [EL]
    spikescounter = 0

x = np.linspace(0, 100, 100)

trace1 = Scatter(x=x, y=variations[0], name= "I = 1")
trace2 = Scatter(x=x, y=variations[1], name= "I = 0.8")
trace3 = Scatter(x=x, y=variations[2], name= "I = 2")
data = [trace1, trace2, trace3]

layout = Layout(
    title='Variation of potential in a neuron with different intensities',
    xaxis=dict(title='Time (ms)'),
    yaxis=dict(title='Electric potential (mV)'))
fig = Figure(data=data, layout=layout)

iplot(fig)
for spikecount in spikecounts:
    print("Number of spikes : " + str(spikecount))
```

A refractory period is a period where the potential is freezed then decreases until it stabilizes to $E_L$. To introduce a refractory period we can implement it like this.

```python
gL = 0.1
EL = -70
I = 1
C = 1
Vs = [EL]
deltat = 1
Vth = -63
refractdelta = 3
refractcount = -1

for i in range(99):
    V = Vs[i]+((gL*(EL-Vs[i])+I)/C)*deltat
    if refractcount == -1:
        Vs.append(V)
        if V >= Vth:
            refractcount += 1
    elif -1 < refractcount < refractdelta:
        Vs.append(Vs[i])
        refractcount += 1
    elif refractcount == refractdelta:
        Vs.append(EL)
        refractcount = -1


x = np.linspace(0, 100, 100)
trace = Scatter(x=x, y=Vs)
data = [trace]

layout = Layout(
    title='Variation of potential in a neuron with refractory period',
    xaxis=dict(title='Time (ms)'),
    yaxis=dict(title='Electric potential (mV)'))
fig = Figure(data=data, layout=layout)

iplot(fig)
```

We observe that when the neuron reached the threshold, it keeps the same value during few ms.

Spike rate adaptation
----

When receiving a constant stimulation, a neuron adapts so that its spike frequency decreases. Let's implement a mechanism fo doing so.

```python
gL = 0.1
EL = -70
EK = -75
I = 5
C = 1
Vs = [EL]
deltat = 1
Vth = -63
deltagsra = 0.04
tausra = 100
gsras = [0]

for i in range(199):
    V = Vs[i]+((gL*(EL-Vs[i])+gsras[i]*(EK-Vs[i])+I)/C)*deltat
    if Vs[i] < Vth:
        Vs.append(V)
        gsras.append(gsras[i]+(-gsras[i]/tausra))
    else:
        Vs.append(EL)
        gsras.append(gsras[i]+deltagsra)

x = np.linspace(0, 200, 200)

trace = Scatter(x=x, y=Vs)
data = [trace]

layout = Layout(
    title='Variation of potential in a neuron with adaptation',
    xaxis=dict(title='Time (ms)'),
    yaxis=dict(title='Electric potential (mV)'))
fig = Figure(data=data, layout=layout)

iplot(fig)
```

As told sooner, the spike frequency diminishes. To understand this, let's plot the adaptation conductance over time.

```python
trace = Scatter(x=x, y=gsras)
data = [trace]

layout = Layout(
    title='Variation of adaptation conductance',
    xaxis=dict(title='Time (ms)'),
    yaxis=dict(title='Adaptation conductance (microS)'))
fig = Figure(data=data, layout=layout)

iplot(fig)
```

Because the neuron fires with a quick rate, the value of $g_{sra}$ increases because the relaxation takes more time than the augmentation with $\Delta g_{sra}$. The adaptation conduction then reaches a threshold and oscillates so that the frequency stabilizes.
